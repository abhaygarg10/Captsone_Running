```python
import numpy as np
import pandas as pd
df = pd.read_csv('../Data-raw/Fertilizer.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
l = ['Rice','Maize','Chickpeas(Channa)','Kidney beans','pigeon peas(Toor Dal)','Moth bean(Matki)','Mung beans','Black Gram','Lentils(Masoor Dal)','Pomegranate','Banana','Mango','Grapes','Water Melon',
'Musk Melon','Apple','Orange','Papaya','Coconut','Cotton','Jute','Coffee']
```

```python
df1 = df.groupby('Crop').mean()
```

```python
df1
```

```python
for i in df1.index:
    if i not in l:
        df1 = df1.drop(i)

df1
```
